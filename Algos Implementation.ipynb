```python
H=B1.x+B0 : Linear regression 



Minimize : 
MSE : (H - y(act))^2

Cost function (loss) = (B1.x+B0 - y)^2

#for optimization, need to minimize B1 and B0
d(loss)/dB1 =0  , d(loss)/dB0 = 0

d(loss)/dB1 = 2(B1.x+B0 - y).x
d(loss)/dB0 = 2(B1.x+B0 - y)

To get to optimal values, B1 and B0 needs to be iterated until minimum value

B1 : B1 - alpha * d(loss)/dB1
B0 : B0 - alpha * d(loss)/dB0
```

```python
from sklearn.datasets import load_diabetes
import pandas as pd
data=load_diabetes()
X=data.data[:,0]
y=data.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
alpha=0.1
B1=0.1
B0=0.1
B1_values=[]
B0_values=[]
error_values=[]
n=len(X_train)
for i in range(n):
    y_pred=B1*X_train[i] + B0
    error = abs(y_train[i]-y_pred)
    
    B1_values.append(B1)
    B0_values.append(B0)
    error_values.append(error)
    
    db1=-2*X[i]*(B1*X_train[i] + B0)
    db0=-2*(B1*X_train[i] + B0)
    
    B1-= db1
    B0-= db0

    
    
```
