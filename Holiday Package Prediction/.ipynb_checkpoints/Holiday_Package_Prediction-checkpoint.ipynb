```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Calculate percentiles
percentile_25 = np.percentile(travel_data['Age'], 25)
percentile_50 = np.percentile(travel_data['Age'], 50)
percentile_75 = np.percentile(travel_data['Age'], 75)

# Plot histogram
sns.histplot(data=travel_data, x='Age', bins=10, kde=True, stat='count')

# Add vertical lines for percentiles
plt.axvline(x=percentile_25, color='g', linestyle='--', label='25th Percentile')
plt.axvline(x=percentile_50, color='b', linestyle='-', label='50th Percentile')
plt.axvline(x=percentile_75, color='r', linestyle='--', label='75th Percentile')

# Display counts on histogram bars
for patch in plt.gca().patches:
    plt.text(patch.get_x() + patch.get_width()/2, 
             patch.get_height(), 
             str(patch.get_height()), 
             fontsize=8, 
             ha='center', 
             va='bottom')

# Show legend and plot
plt.legend(title='Percentiles')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
travel_data['Gender'] = travel_data['Gender'].replace('Fe Male', 'Female')
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = travel_data.select_dtypes(include=['object', 'category']).columns
ohe = OneHotEncoder(sparse_output=False)
travel_data_ohe = ohe.fit_transform(travel_data[categorical_cols])
travel_data_ohe = pd.DataFrame(travel_data_ohe, columns=ohe.get_feature_names_out())

# # Concatenate with non-categorical columns
non_categorical_cols = travel_data.select_dtypes(exclude=['object', 'category']).columns
travel_data_final = pd.concat([travel_data[non_categorical_cols], travel_data_ohe], axis=1)
```
